```python
def relative(column, data):
    a = column.value_counts()
    data['a'] = column.map(a)
    print(data['a'].isna().sum())
    b = column[data['goal1'] == 1].value_counts()
    data['b'] = column.map(b)
    data['b'] = data['b'].replace(np.NaN, 0)
    perc = data['b']/data['a']
    data = data.drop(['a','b'], axis=1, inplace=True)
    return perc
```
